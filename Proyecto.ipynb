## 1. TOKENIZACION

```python
# Asegurarse de que la columna 'desc' no tenga valores nulos o no textuales
desc_col = desc_col.dropna()  # Elimina valores nulos
desc_col = desc_col[desc_col.apply(lambda x: isinstance(x, str))]  # Mantener solo cadenas de texto

print(f"Filas válidas para tokenización: {len(desc_col)}")

```

```python
# Tokenización
tokens_solution = []

for n, text in enumerate(desc_col):
    if not n % 100:
        print(f'\rTokenizing data #{n} out of {len(desc_col)}', end='', flush=True)
    try:
        # Tokenizamos cada entrada de texto
        tokens = word_tokenize(text)
        tokens_solution.append(tokens)
    except Exception as e:
        print(f"\nError tokenizing row {n}: {text}")
        print(e)
        tokens_solution.append([])  # Agregar una lista vacía si hay un error

print('\n\nThe data has been tokenized. Check the result for the first entry:')
print(desc_col.iloc[0])  # Primera entrada del texto original
print(tokens_solution[0])  # Tokens correspondientes
```

```python
#-------- TOKENIZACION --------

length = len(desc_col)
tokens_solution = []

for n, text in enumerate(desc_col):
    if not n%100:
        print('\rTokenizing data #', n, 'out of', length, end='', flush=True)

    # Tokenizamos cada entrada de texto
    tokens = word_tokenize(text) #esto es una lista de palabras o caracteres (tokens) que ha seleccionado

    # creamos una lista de listas (añadiendo cada token)
    tokens_solution.append(tokens)

print('\n\n The data has been tokenized. Check the result for the first cell of the summary col:')
print(desc_col[0])
print(tokens_solution[0])
```
